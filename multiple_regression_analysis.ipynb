Последний шаг в подготовке данных: линейная регрессия в python не справляется с категориальными переменными (типом object в pandas), поэтому давайте применим pd.get_dummies(). 

```python
df_dummy = pd.get_dummies(data=cars_1[['fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginetype', 'cylindernumber', 'company']], drop_first = True)
```

```python
df = pd.concat([cars_2, df_dummy], axis=1)
```

Сначала построим небольшую модель всего с одним предиктором цены (price) – horsepower.
Какой процент изменчивости объясняет полученная модель? 

```python
# способ первый
# Y = одномерный массив с ЗП, X - массив со всеми нужными нам НП
Y = df.price
X = sm.add_constant(df.horsepower) # добавить константу, чтобы был свободный член
model = sm.OLS(Y, X) # говорим модели, что у нас ЗП, а что НП
results = model.fit() # строим регрессионную прямую
print(results.summary()) # смотрим результат

#получается что мы в 65% случаев верно определим цену машины по одному параметру. 
#Есть более значимые и менее значимые параметры 
```
